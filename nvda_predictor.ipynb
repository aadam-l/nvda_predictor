```python
# Yahoo Finance API.
import yfinance as yf
```

```python
# Initialize a ticker class to download price history for a specified symbol.
nvda = yf.Ticker("NVDA")
```

```python
# Get historical market data.
nvda = nvda.history(period="max")
```

```python
nvda
```

```python
nvda.index
```

```python
# Plot data using the index and closing price.
nvda.plot.line(y="Close", use_index=True)
```

```python
# Removing the dividends and stock splits column.
del nvda["Dividends"]
del nvda["Stock Splits"]
```

```python
# Pandas shift() function.
nvda["Tomorrow"] = nvda["Close"].shift(-1)
```

```python
nvda
```

```python
# Return 1 if the tomorrow price > close price (a price increase)
# or 0 if the tomorrow price < close price (a price decrease).
nvda["Target"] = (nvda["Tomorrow"] > nvda["Close"]).astype(int)
```

```python
nvda
```

```python
# Utilize a RandomForestClassifier, which trains individual decision trees with
# randomized parameters and averages their results. This method offers greater resistance
# to overfitting compared to other data models.
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = nvda.iloc[:-100]
test = nvda.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
# What percentage of our market predictions for an upward trend were accurate?
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
# Plot predictions with actual values
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(nvda, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# Mean closing price in the last 2 days, trading week, 3 months, year, and 4 years
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = nvda.rolling(horizon).mean()

    # Column calculating the ratio of today's closing price to the rolling average
    # closing price over the last 2 days, trading week, 3 months, year, and 4 years
    ratio_column = f"Close_Ratio_{horizon}"
    nvda[ratio_column] = nvda["Close"] / rolling_averages["Close"]

    # Column counting the number of days the stock price increased over the past x days
    trend_column = f"Trend_{horizon}"
    nvda[trend_column] = nvda.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
# Removing rows with missing data
nvda = nvda.dropna()
```

```python
nvda
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    # Returns a probability that the row returns a 1 or 0
    preds = model.predict_proba(test[predictors]) [:,1]
    # Creating a custom threshold
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(nvda, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
